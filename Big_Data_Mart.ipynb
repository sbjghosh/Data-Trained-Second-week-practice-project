```python
x=data.drop(['Item_Outlet_Sales'],axis=1)
y=data['Item_Outlet_Sales']
```

```python
from sklearn import model_selection
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
X_train,X_test,y_train,y_test=model_selection.train_test_split(x,y,test_size=0.25)
```

```python
linear_pipeline=Pipeline([('Normalization',StandardScaler()),('Linear Regression',LinearRegression())])
linear_pipeline.fit(X_train,y_train)
metrics.r2_score(linear_pipeline.predict(X_test),y_test)
```

```python
rf_pipeline=Pipeline([('Normalization',StandardScaler()),('RF',RandomForestRegressor())])
rf_pipeline.fit(X_train,y_train)
metrics.r2_score(rf_pipeline.predict(X_test),y_test)
```
